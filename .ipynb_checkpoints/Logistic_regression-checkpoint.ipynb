```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.decomposition import PCA
from sklearn.manifold import TSNE
```

```python
from sklearn.datasets import load_breast_cancer
```

```python
cancer = load_breast_cancer()
```

```python
df = pd.DataFrame(
    data = cancer['data'], columns = cancer['feature_names']
)
```

```python
df['target'] = cancer['target']
```

```python
df
```

```python
X = cancer.data
y = cancer.target
```

```python
pca = PCA(n_components=2)
tsne = TSNE(n_components = 2)
```

```python
pca_X = pca.fit_transform(X)
tsne_X = tsne.fit_transform(X)
```

```python
df = pd.DataFrame(data = pca_X, columns = ['pca1', 'pca2'])
df['target'] = y
df2 = pd.DataFrame(data = tsne_X, columns = ['tsne1', 'tsne2'])
df2['target'] = y
```

```python
df.head(), df2.head()
```

```python
plt.scatter(df[df['target'] == 0]['pca1'], df[df['target'] == 0]['pca2'], color= 'red')
plt.scatter(df[df['target'] == 1]['pca1'], df[df['target'] == 1]['pca2'], color = 'blue')
```

```python
plt.scatter(df2[df2['target'] == 0]['tsne1'], df2[df2['target'] == 0]['tsne2'], color= 'red')
plt.scatter(df2[df2['target'] == 1]['tsne1'], df2[df2['target'] == 1]['tsne2'], color = 'blue')
```
